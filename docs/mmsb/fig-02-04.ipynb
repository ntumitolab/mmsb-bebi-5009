# Fig 2.04 Exponential decay

```julia
using Plots
Plots.gr(lw=2)
```

```julia
plot([t-> 3 * exp(-t) t->3 * exp(-2t) t-> 3 * exp(-3t)], 
     0.0, 5.0, 
     xlim = (0, 5), ylim=(0, 3.2),
     xlabel="Time", ylabel="Concentration", 
     label = ["exp(-t)" "exp(-2t)" "exp(-3t)"], 
     title= "Figure 2.4")
```
